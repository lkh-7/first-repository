유니코드는 나라마다 다른 문자,숫자등을 컴퓨터가 올바르게 인식하도록 ISO에서 맺은 약속이다. 미국만의 표준인 ascii코드에서, 전세계적으로 맺은 약속이고, UTF-8은 유니코드를 사용하븐 방시으로 알고 있다, UTF-8은 1byte~4byte까지 적은양의 문자부터 많은 양의 문자까지 포함해서 사용가능하고, UTF-18은 2byte,4byte로 큰양의 문자들을 취급하는 걸로 알고 있다.
파이썬은 파이썬3.0부터 UTF-8을 표준으로 채택해서 사용하고 있다.
유니코드를 만들어볼 전문적인 고부가 아닌 이상, 이런게 있다는 것정도만 알아도 될것 같다.

 양이 많은 문서에서, 어떤 단어를 찾아야할때, 끝몇개글자가 동일한 단어가 많다면, endswith()함수가 유용하고, 첫몇개글자가 동일한 단어가 많다면,startswith()가 유용 
 
 #### - LMS 에 링크된 자료를 참고하여 mutable 과 immutable 의 차이점을 설명해보았다.
 mutable는 데이터구조상 컨테이너역할을 하는 리스트,dict,set은 내부의 요소를 교체해도 되며, 교체후에도 메모리주소는 동일함, 이렇게 교체를 해도 메모리번호가 동일한 데이터를 mutable하다고 함.
